```python
#POVIT_TABLE
Book_df=Book1.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating')
```

```python
#CALCULATING THE COSINE SIMILARITY
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_similarity=1-pairwise_distances(Book_df.values,metric='cosine')
```

```python
user_similarity
```

```python
user_sim_df=pd.DataFrame(user_similarity)
```

```python
user_sim_df.index=Book1.User_ID.unique()
user_sim_df.columns=Book1.User_ID.unique()
```

```python
def get_record(User_ID):
    try:
        tem=list(user_sim_df.sort_values([User_ID],ascending=False).head(100).index)
   
        book_list=list(Book1[Book1["User_ID"].isin(tem)]["Book_Title"])
        #print(book_list)
        return set(book_list)-set(Book1[Book1["User_ID"]==User_ID]["Book_Title"])
    
    except:
        return "the user id isn't available"
```
